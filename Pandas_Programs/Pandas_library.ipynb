3. Write a Python program to add, subtract, multiple and divide two Pandas Series.
Sample Series: [2, 4, 6, 8, 10], [1, 3, 5, 7, 9]

```python


def perform_operations(series1, series2):
    
    
    '''
    Description :
        This function performs addition, subtraction, multiplication, and division 
        on two provided Pandas Series and prints the results.
    
    Parameters :
        series1 : pd.Series : The first Pandas Series
        series2 : pd.Series : The second Pandas Series
    
    Returns :
        None
    '''    
    
    
    addition = series1 + series2
    print(f"Addition:\n{addition}\n")

    subtraction = series1 - series2
    print(f"Subtraction:\n{subtraction}\n")

    multiplication = series1 * series2
    print(f"Multiplication:\n{multiplication}\n")

    
    division = series1 / series2
    print(f"Division:\n{division}\n")



def main():

    series1 = pd.Series([2, 4, 6, 8, 10])
    series2 = pd.Series([1, 3, 5, 7, 9])

    print(f"Series 1:\n{series1}\n")
    print(f"Series 2:\n{series2}\n")
    
    perform_operations(series1, series2)

if __name__ == "__main__":
    main()
```

4. Write a Python program to get the powers of an array values element-wise.
Note: First array elements raised to powers from second array

```python

def square_elements(arr):
    
    
    '''
    Description :
        This function squares each element in the provided Pandas array.
    
    Parameters :
        arr : pd.Array : The Pandas array whose elements need to be squared
    
    Returns :
        pd.Array : The modified Pandas array with squared elements
    '''   
    
     
    for index in range(len(arr)):
        arr[index] = arr[index] ** arr[index]
    return arr



def main():
    

    arr_1 = pd.array([2, 4, 6, 8, 10])
    print(f"Original Array:\n{arr_1}\n")
    
    squared_arr = square_elements(arr_1)
    print(f"Squared Array:\n{squared_arr}\n")

if __name__ == "__main__":
    main()
```

22. Write a Python program to iterate over rows in a DataFrame.

```python

def process_exam_data(df):
    
    
    """
    Description:
        This function processes the exam DataFrame by converting each row into a list
        and appending it to a list of rows.
    
    Parameter:
        df (pd.DataFrame): The DataFrame containing exam data.
    
    Return:
        row_list (list): A list of lists, where each sub-list represents a row from the DataFrame.
    """
    
    
    row_list = []

    for index in range(len(df)):
        row_list.append(df.iloc[index])

    print(df)
    print()
    print(row_list)
    


def main():

   
    exam_data = {
        'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael',
                 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
    }
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

  
    df = pd.DataFrame(data=exam_data, index=labels)

  
    process_exam_data(df)


if __name__ == "__main__":
    main()
```
