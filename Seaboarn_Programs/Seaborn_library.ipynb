<h1>Seaborn Bar plot<h1>

1. Write a program to draw bar plot of sex against survived for a dataset given in the url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python

def load_data(url):
    
    
    """
    Description:
        Loads the Titanic dataset from the given URL.
        
    Parameter:
        url (str): The URL to the Titanic dataset.
        
    Returns:
        pd.DataFrame: The loaded Titanic dataset as a DataFrame.
    """
    
    return pd.read_csv(url)



def plot_sex_survived(df):
    
    
    
    """
    
    Description:
        Generates a bar plot of sex against survived using the provided DataFrame.
        
    Parameter:
        df (pd.DataFrame): The DataFrame containing Titanic data.
        
    Returns:
        None
        
    """
    
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(8, 6))
    sns.countplot(x='sex', hue='survived', data=df)
    plt.title("Bar Plot of Sex Against Survived")
    plt.xlabel("Sex")
    plt.ylabel("Count")
    plt.legend(title='Survived', loc='upper right')
    plt.show()



def main():

    url = "https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv"
    df = load_data(url)
    plot_sex_survived(df)

if __name__ == "__main__":
    main()
```

<h1 >Seaborn BoxPlot <h1>

1. Write a program to draw box plot of life expectancy by continent for a data set given in a
url
https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapmind
er-FiveYearData.csv

```python


def load_data(url):
    
    
    """
    Description:
        Loads the Gapminder dataset from the given URL.
        
    Parameter:
        url (str): The URL to the Gapminder dataset.
        
    Returns:
        pd.DataFrame: The loaded Gapminder dataset as a DataFrame.
    """
    
    
    return pd.read_csv(url)



def plot_life_expectancy_box(df):
    
    
    """
    Description:
        Generates a box plot of life expectancy by continent using the provided DataFrame.
        
    Parameter:
        df (pd.DataFrame): The DataFrame containing Gapminder data.
        
    Returns:
        None
    """
    
    
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='continent', y='lifeExp', data=df)
    plt.title("Box Plot of Life Expectancy by Continent")
    plt.xlabel("Continent")
    plt.ylabel("Life Expectancy")
    plt.show()



def main():

    url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
    df = load_data(url)
    plot_life_expectancy_box(df)

if __name__ == "__main__":
    main()
```
