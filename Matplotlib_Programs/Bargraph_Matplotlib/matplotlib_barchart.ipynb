1. Write a Python programming to display a bar chart of the popularity of programming
Languages.
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python

def barchart_popularity(languages,popularity):

    """
    Description:
        Function will isplay a bar chart of the popularity of programming Languages.

    Parameter:
        languages:list of languages
        popularity:list of popularity of language respectfully
      
    Return:
        None    
    """

    plt.figure(figsize=(10, 6))
    plt.bar(languages, popularity)

    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity')

    plt.show()



def main():

    languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
    barchart_popularity(languages,popularity)
 

if __name__=="__main__":
    main()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)

```python

def horizontal_gender_barchart(means_men,means_women):

    """
    Description:
        Function will display a bar chart of the popularity of programming Languages in horizontal and increase bottom margin

    Parameter:
        means_men:mens score in list
        means_women:women scores in list
      
    Return:
        None    
    """

    groups = ['Group1', 'Group2', 'Group3', 'Group4', 'Group5']

    n_groups = len(groups)

    index = np.arange(n_groups)

    bar_width = 0.2

    plt.bar(index, means_men, bar_width, color='red', label='Men')

    plt.bar(index + bar_width, means_women, bar_width, color='yellow', label='Women')

    plt.xlabel('Groups')
    plt.ylabel('Scores')
    plt.title('Scores by group and gender')
    plt.xticks(index + bar_width / 2, groups) 
    plt.legend()

    plt.show()


def main():
    try:
        means_men = [22, 30, 35, 35, 26]
        means_women = [25, 32, 30, 35, 29]
        horizontal_gender_barchart(means_men,means_women)


    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

12. Write a Python program to create bar plots with error bars on the same figure. Sample Date
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python


def error_barchart(mean_velocity,std_deviation):

    """
    Description:
        Function will create bar plot with error bars on same figure

    Parameter:
        mean_velocity:means value
        std_deviation:standard deviation value
      
    Return:
        None    
    """

    categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']

    x = np.arange(len(categories))

    bar_width = 0.5

    plt.bar(x, mean_velocity, bar_width, yerr=std_deviation, color='skyblue', edgecolor='black', label='Mean Velocity')

    plt.xlabel('Categories')
    plt.ylabel('Mean Velocity')
    plt.title('Bar Plot with Error Bars')
    plt.xticks(x, categories) 
    plt.legend()

    plt.show()


def main():
    try:
        mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
        std_deviation = [0.3314, 0.2278, 0.2836, 0.2645]
        error_barchart(mean_velocity,std_deviation)

    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()
```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text
label above each bar displaying men means (integer value).
Sample Data
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python

def error_of_barchart(mean_velocity,standard_deviation):

    """
    Description:
        Function will create bar plot with error bars on same figure  will  add text label

    Parameter:
        mean_velocity:means value
        std_deviation:standard deviation value
      
    Return:
        None    
    """

    categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
    mean_velocity_int = [int(round(value  * 1000)) for value in mean_velocity]

    x = np.arange(len(categories))

    bar_width = 0.5

    plt.figure(figsize=(10, 6))
    bars = plt.bar(x, mean_velocity, bar_width, yerr=standard_deviation, capsize=5, color='skyblue', edgecolor='black', label='Mean Velocity')

    plt.xlabel('Categories')
    plt.ylabel('Mean Velocity')
    plt.title('Bar Plot with Error Bars and Text Labels')
    plt.xticks(x, categories)  
    plt.legend()


    for bar, mean in zip(bars, mean_velocity_int):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{mean}', ha='center', va='bottom')

    plt.show()


def main():

    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    standard_deviation = [0.3314, 0.2278, 0.2836, 0.2645]
    error_of_barchart(mean_velocity,standard_deviation)

if __name__=="__main__":
    main()
```
