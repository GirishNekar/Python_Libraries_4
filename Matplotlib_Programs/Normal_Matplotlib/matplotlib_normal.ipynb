3. Write a Python program to draw a line using given axis values taken from a text file, with
suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2
2 4
3 1

```python

def draw_line(xpoints,ypoints):

    """
    Description:
        function will plot the straight line graph using matplotlib plot 

    Parameter:
        array1,array2:will use this this array as coordinates to plot graph

    Return:
        None    
    """
    
    plt.plot(xpoints,ypoints)
    plt.xlabel("x axis")
    plt.ylabel("y axis")
    plt.title("Graph",loc="center")
    plt.show()
   

def main():
    array1=[]
    array2=[]

    with open('test.txt','r') as file:
            
        for line in file:
            values=line.split()
            if  len(values)==2:
                if values[0].isdigit() and values[1].isdigit():
                    array1.append((int)(values[0]))
                    array2.append((int)(values[1]))

    draw_line(np.array(array1),np.array(array2))


if __name__=="__main__":
    main()
```

5. Write a Python program to plot two or more lines on same plot with suitable legends of each
line.

```python


def plot_two_line(ypoints1,ypoints2):

    """
    Description:
        Function will plot two or more lines on same plot 

    Parameter:
       ypoints1:y axis points in array1
       ypoints2:y axis points in array2

    Return:
        None    
    """
    
    plt.plot(ypoints1,color='hotpink',ls=':',label="first plot")
    plt.plot(ypoints2,color='red',ls='--',label="second plot")
    plt.xlabel("X axis")
    plt.ylabel("Y axis")
    plt.title("Graph with two or more plot")
    plt.legend()
    plt.show()


def main():

    ypoints1=np.random.randint(0,8,size=8)
    ypoints2=np.random.randint(0,8,size=8)
        
    plot_two_line(ypoints1,ypoints2)

if __name__=="__main__":
    main()
```

9. Write a Python program to display the current axis limits values and set new axis values.

```python

def plot_line(ypoints):

    """
    Description:
        Function will display the current axis limits values and set new axis values.

    Parameter:
       ypoints:y axis points in an array
      
    Return:
        None    
    """
    
    plt.plot(ypoints,color='hotpink',ls=':',marker='o')

    current_x_limits = plt.xlim() 
    current_y_limits = plt.ylim()  

    print("Current x-axis limits:", current_x_limits)
    print("Current y-axis limits:", current_y_limits)
    
    plt.xlabel("X axis")
    plt.ylabel("Y axis")
    plt.title("Line Plot with Custom Axis Limits")
    plt.xlim(2, 20) 
    plt.ylim(2, 20) 

    new_x_lim = plt.xlim()  
    new_y_lim = plt.ylim()  

    print("New x-axis limits:", new_x_lim)
    print("New y-axis limits:", new_y_lim)

    plt.show()


def main():
    
    ypoints=np.random.randint(0,20,size=10)

    plot_line(ypoints)
   

if __name__=="__main__":
    main()
```

11. Write a Python program to plot several lines with different format styles in one command
using arrays.

```python


def plot_multiple_line():

    """
    Description:
        Function will plot several lines with different format styles in one command
        using arrays.

    Parameter:
        None
      
    Return:
        None    
    """
    xpoint = np.linspace(0, 10, 100) 

    ypoint1 = np.sin(xpoint)        
    ypoint2 = np.cos(xpoint)          
    ypoint3 = np.tan(xpoint)          

    plt.plot(xpoint, ypoint1, 'r--', xpoint, ypoint2, 'g-.', xpoint, ypoint3, 'b:')

    plt.xlabel('X Axis')
    plt.ylabel('Y Axis')
    plt.title('Multiple Lines with Different Format Styles')

    plt.legend(['Sine_wave', 'Cos_wave', 'Tan_wave'])

    plt.show()


def main():
    plot_multiple_line()
    
if __name__=="__main__":
    main()
```

12. Write a Python program to create multiple types of charts

```python


def plot_multiple_type_chart():

    """
    Description:
        Function will create multiple types of charts.
        using arrays.

    Parameter:
        None
      
    Return:
        None    
    """
    xpoints = np.linspace(0, 10, 100)
    ypoints1 = np.sin(xpoints)
    ypoints2 = np.cos(xpoints)
    categories = ['A', 'B', 'C', 'D', 'E']
    values = [10, 20, 15, 25, 30]
    sizes = [15, 30, 45, 10,20]


    fig, axs = plt.subplots(1, 3, figsize=(20,10 ))

    axs[0].plot(xpoints, ypoints1, 'r--', label='Sine')
    axs[0].plot(xpoints, ypoints2, 'b-.', label='Cosine')
    axs[0].set_title('Line Chart')
    axs[0].set_title('Line Chart')
    axs[0].set_xlabel('X Axis')
    axs[0].set_ylabel('Y Axis')
    axs[0].legend()

    axs[1].bar(categories, values, color='green')
    axs[1].set_title('Bar Chart')
    axs[1].set_title('Bar Chart')
    axs[1].set_xlabel('Categories')
    axs[1].set_ylabel('Values')

    axs[2].pie(sizes, labels=categories, autopct='%1.1f%%', startangle=140)
    axs[2].set_title('Pie Chart')

    plt.tight_layout()
    plt.show()


def main():
    try:
        plot_multiple_type_chart()


    except ValueError as e:
        print(f"Invalid input -->{e}")    

if __name__=="__main__":
    main()

```

14. Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
off ticks.

```python



def plot_chart_with_grid():

    """
    Description:
        Function will display the grid and draw line charts of the closing value of
        Alphabet Inc. between October 3, 2016 to October 7, 2016. will turn on grid but turn off ticks

    Parameter:
        None
      
    Return:
        None    
    """

    data_frame = pd.read_csv("new_fdata.csv", parse_dates=['Date'])

    plt.figure(figsize=(10, 6))
    plt.plot(data_frame['Date'], data_frame['Close'], marker='o', color='b', linestyle='-', label='Closing Value')

    plt.grid(which='both', linestyle='-', linewidth=0.5, color='blue')  # Major grid
    plt.minorticks_on()  
    plt.grid(which='minor', linestyle='-.', linewidth=0.5, color='cyan')  # Minor grid

    
    plt.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False)

    plt.title('Closing Value of Alphabet Inc. (Oct 3, 2016 - Oct 7, 2016)')
    plt.xlabel('Date')
    plt.ylabel('Closing Value')

    plt.legend()

    plt.tight_layout()
    plt.show()


def main():
    plot_chart_with_grid()


if __name__=="__main__":
    main()
```

15. Write a Python program to create multiple plots.

```python


def multiple_plot():

    """
    Description:
        Function will create multiple plots.

    Parameter:
        None
      
    Return:
        None    
    """

    x = np.linspace(0, 2 * np.pi, 100)  
    y_sin = np.sin(x) 
    y_cos = np.cos(x)  
    y_tan = np.tan(x)  


    plt.figure(figsize=(10, 20))

    plt.subplot(3, 1, 1)
    plt.plot(x, y_sin, color='r', linestyle='-', marker='o', label='Sine')
    plt.title('Sine Function')
    plt.xlabel('X Axis')
    plt.ylabel('sin(x)')
    plt.legend()
    plt.grid(True)

    plt.subplot(3, 1, 2)
    plt.plot(x, y_cos, color='b', linestyle='--', marker='x', label='Cosine')
    plt.title('Cosine Function')
    plt.xlabel('X Axis')
    plt.ylabel('cos(x)')
    plt.legend()
    plt.grid(True)

    plt.subplot(3, 1, 3)
    plt.plot(x, y_tan, color='g', linestyle='-.', marker='s', label='Tangent')
    plt.title('Tangent Function')
    plt.xlabel('X Axis')
    plt.ylabel('tan(x)')
    plt.ylim(-10, 10)  
    plt.legend()
    plt.grid(True)

  
    plt.tight_layout()


    plt.show()


def main():
    
    multiple_plot()  

if __name__=="__main__":
    main()
```
