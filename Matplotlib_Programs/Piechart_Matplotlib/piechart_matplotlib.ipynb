1. Write a Python programming to create a pie chart of the popularity of programming
Languages.

```python



def plot_pie_chart(prog_lang, popularity):
    
    
    """
    Description: Plots a pie chart representing the popularity of programming languages.
    
    Parameters:
    prog_lang (list): A list of programming languages.
    popularity (list): A list of popularity values corresponding to the programming languages.
    
    Returns:
    None
    """
      
    plt.pie(popularity, labels=prog_lang, autopct='%1.1f%%', startangle=90)        
    plt.show()


def initialize_data():
    
    
    """
    Description: Initializes the data for programming languages and their popularity.
    
    Parameters: None
    
    Returns:
    tuple: A tuple containing two lists: 
           - prog_lang: List of programming languages.
           - popularity: List of popularity values.
    """
    
    
    prog_lang = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]
    return prog_lang, popularity


def main():

    prog_lang, popularity = initialize_data()
    plot_pie_chart(prog_lang, popularity)

if __name__ == "__main__":
    main()
```

3 . Write a Python programming to create a pie chart of gold medal achievements of five most
successful countries in 2016 Summer Olympics. Read the data from a csv file.

```python


def plot_pie_chart_from_csv(file_path):
    
    """
    Description: Reads data from a CSV file and plots a pie chart of gold medal achievements.
    
    Parameters:
    file_path (str): The path to the CSV file containing the data.
    
    Returns:
    None
    """

    data = pd.read_csv(file_path)
        
    countries = data['country']
    gold_medals = data['gold_medal']
        
      
    plt.figure(figsize=(8, 8))
    plt.pie(gold_medals, labels=countries, autopct='%1.1f%%', startangle=140)  
    plt.title('Gold Medal Achievements in 2016 Summer Olympics')    
    plt.show()

def main():
 
    file_path = 'olympic_data.csv' 
    plot_pie_chart_from_csv(file_path)

if __name__ == "__main__":
    main()
```
